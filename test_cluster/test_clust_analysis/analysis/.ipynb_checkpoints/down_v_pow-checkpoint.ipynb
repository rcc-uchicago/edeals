```python
%matplotlib inline
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
plt.style.use('willo')
```

```python
strux = pd.read_csv('../../data/data_exp_strux/strux.csv',thousands=',')
strux['Time']=pd.to_datetime(strux['Time'])
```

```python
noffs = range(7)
offsets = [ pd.Timedelta('0 hours'),  pd.Timedelta('4 hours'),  pd.Timedelta('8 hours'),  pd.Timedelta('12 hours')]

cdut = []
cdup = []
cduo = []
ipmit = [] 
ipmip = [] 
ipmio = [] 
percent = []
for t_offset in offsets:
    dcdut=[]
    dcdup=[]
    dcduo=[]
    dipmit=[]
    dipmip=[]
    dipmio=[]
    dpercent =[]
    for noff in noffs:
        gang = pd.read_csv('../../data/data_exp_'+str(noff)+'/gang_data.csv')
        gang['Timestamp']=pd.to_datetime([val.rsplit('-',1)[0] for val in gang['Timestamp']])
        t_dn = gang['Timestamp'].iloc[0]+pd.Timedelta('15 minutes') + t_offset
        t_on = gang['Timestamp'].iloc[0]+pd.Timedelta('45 minutes') + t_offset
        t_end = gang['Timestamp'].iloc[0]+pd.Timedelta('3.5 hours') + t_offset

        struxt = strux[(strux['Time']>t_dn) & (strux['Time']<t_end)]
        struxp = strux[(strux['Time']>t_dn) & (strux['Time']<t_on)]
        struxo = strux[(strux['Time']>t_on) & (strux['Time']<t_end)]

        cdu_dt = (struxt['Time'].iloc[1]-struxt['Time'].iloc[0]).seconds/3600.
        cdu_powt = cdu_dt*np.sum(struxt['Value'].astype('int').values)
        cdu_powp = cdu_dt*np.sum(struxp['Value'].astype('int').values)
        cdu_powo = cdu_dt*np.sum(struxo['Value'].astype('int').values)

        strt = 5
        todo = 6

        ipmi_powt = 0
        ipmi_powp = 0
        ipmi_powo = 0
        mn = 0
        for index in range(strt,strt+todo):

            pick = 'midway'+str(index).zfill(3)
            sub = pick+'|Timestamp'  
            subgang = gang.filter(regex=sub)
            subgang.columns=["".join(col.lstrip(pick).split()) for col in subgang.columns]
            ipmi_dt = (subgang['Timestamp'][1]-subgang['Timestamp'][0]).seconds/3600.
            if('domain_avg_power' in subgang):
                gangt = subgang[(subgang['Timestamp']>t_dn) & (subgang['Timestamp']<t_end)]
                gangp = subgang[(subgang['Timestamp']>t_dn) & (subgang['Timestamp']<t_on)]
                gango = subgang[(subgang['Timestamp']>t_on) & (subgang['Timestamp']<t_end)]

                ipmi_powt = ipmi_powt + ipmi_dt*np.sum(gangt['domain_avg_power'])
                ipmi_powp = ipmi_powp + ipmi_dt*np.sum(gangp['domain_avg_power'])
                ipmi_powo = ipmi_powo + ipmi_dt*np.sum(gango['domain_avg_power'])
                mn = mn+ np.mean(gango['proc_run'])
        dpercent.append(mn/todo)    
        dcdut.append(cdu_powt)
        dcdup.append(cdu_powp)
        dcduo.append(cdu_powo)
        dipmit.append(ipmi_powt)
        dipmip.append(ipmi_powp)
        dipmio.append(ipmi_powo)
    percent.append(dpercent)
    cdut.append(dcdut)
    cdup.append(dcdup)
    cduo.append(dcduo)
    ipmit.append(dipmit)
    ipmip.append(dipmip)
    ipmio.append(dipmio)
percent = np.array(percent)/12
cdut = np.array(cdut)/1000
cdup = np.array(cdup)/1000
cduo = np.array(cduo)/1000

ipmit = np.array(ipmit)/1000
ipmip = np.array(ipmip)/1000
ipmio = np.array(ipmio)/1000
```

```python
mean_wait = []
max_wait = []
last_job = []
for t_offset in offsets:
    dmean_wait = []
    dmax_wait = []
    dmean_run = []
    dmax_run = []
    dlast_job = []
    for noff in noffs:
        slurm = pd.read_csv('../../data/data_exp_'+str(noff)+'/slurm_data.csv',header=None)
        slurm[1]=pd.to_datetime(slurm[1])
        slurm[2]=pd.to_datetime(slurm[2])
        slurm[21]=pd.to_timedelta(slurm[21])

        t_str = slurm[1].min() + t_offset
        t_end = slurm[1].min() + pd.Timedelta('3.75 hours') + t_offset
        
        strt = 5
        todo = 6
        inds = slurm[0]=='p'
        for index in range(strt,strt+todo):
            pick = 'midway'+str(index).zfill(3)
            inds = inds | (slurm[0]==pick)
        inds = inds & (slurm[1]>t_str) & (slurm[1]<t_end)
        subslurm = slurm[inds]
        dmean_wait.append(subslurm[21].mean().seconds+subslurm[21].mean().microseconds/1000000.)
        dmax_wait.append(subslurm[21].max().seconds+subslurm[21].max().microseconds/1000000.)
        dmean_run.append((subslurm[2]-subslurm[1]).mean().seconds+(subslurm[2]-subslurm[1]).mean().microseconds/1000000.)
        dmax_run.append((subslurm[2]-subslurm[1]).max().seconds+(subslurm[2]-subslurm[1]).max().microseconds/1000000.)
        dlast_job.append((subslurm[1].max()-subslurm[1].min()).seconds)
    mean_wait.append(dmean_wait)
    max_wait.append(dmax_wait)
    last_job.append(dlast_job)
mean_wait = np.array(mean_wait)/60.
max_wait = np.array(max_wait)/60.
last_job = np.array(last_job)/3600.
```

```python
f = plt.figure()
ax = plt.gca()
# co = next(ax._get_lines.color_cycle)
# plt.plot(noffs,cdut/3.25,label='cdu total',color=co)
# plt.plot(noffs,ipmit/3.25,'--',label='ipmi total',color=co)
labs = ['75%','65%','60%','55%']
noffs = np.array(noffs)
for ind in range(4):
    co = next(ax._get_lines.color_cycle)
    plt.plot(noffs/6.,cdup[ind]/0.5,label=labs[ind],color=co)
#     plt.plot(noffs,ipmip[ind]/0.5,'--',color=co)
#     co = next(ax._get_lines.color_cycle)
    plt.plot(noffs/6.,cduo[ind]/2.75,'--',color=co)
#     plt.plot(noffs,ipmio[ind]/2.75,'--',color=co)
plt.ylim([0,1.5])
plt.ylabel('Energy Usage (kWh/hr)')
plt.xlabel('Fraction Shutdown')
plt.legend(title='Mean Cluster Use',loc=3)
f.savefig("../figures/down_v_pow.pdf", format='pdf', dpi=1000)

f2  = plt.figure()
ax.set_color_cycle(None)
for ind in range(4):
    co = next(ax._get_lines.color_cycle)
    plt.plot(noffs/6.,max_wait[ind],label=labs[ind],color=co)
    plt.plot(noffs/6.,mean_wait[ind],'--',color=co)
plt.ylabel('Job Wait Time (min)')
plt.xlabel('Fraction Shutdown')
plt.legend(title='Mean Cluster Use',loc=3)
f2.savefig("../figures/down_v_wait.pdf", format='pdf', dpi=1000)

```

```python
import scipy.optimize as optim

def model_func(x,u0,uv):
    return u0+uv*x

ip = np.array(1000*ipmip[:]/0.5/6).flatten()

cd = np.array(1000*cdup[:]/0.5/6).flatten()

c_params = optim.curve_fit(model_func,ip,cd,np.array([0.0,0.0]))

f3  = plt.figure()
ax.set_color_cycle(None)
plt.scatter(ip,cd, label='Test Cluster Measurements')
plt.scatter(1000*ipmio[:]/2.75/6,1000*cduo[:]/2.75/6)
rng = np.linspace(0,200,50);
plt.plot(rng,c_params[0][0]+c_params[0][1]*rng,lw=3,c='#ec6778',label="Best Fit")
plt.legend(loc=2)
plt.xlabel('IPMI Avg Power Measurement (W)')
plt.xlim([0,200])
plt.ylabel('CDU Avg Power Measurement (W)')
f3.savefig("../figures/ipmi_v_cdu.pdf", format='pdf', dpi=1000)
print str(c_params[0][0]) + " " + str(c_params[0][1])
```
